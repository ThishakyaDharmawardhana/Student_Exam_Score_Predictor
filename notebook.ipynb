1️⃣ What train_test_split does

- train_test_split is a function from scikit-learn that splits your dataset into training and testing sets.
- Training set → used to train the model
- Testing set → used to evaluate the model (check how well it predicts unseen data)

2️⃣ Inputs

- X → the features/input variables (all the columns you want to use to predict)
- y → the target/output variable (what you want to predict)
- test_size=0.2 → 20% of the data will be used as the test set, and 80% for training

3️⃣ Outputs

The function returns 4 things:

Variable	              Meaning
 x_train	     Features for training (80% of X)
 x_test	         Features for testing (20% of X)
 y_train	     Target for training (80% of y)
 y_test	         Target for testing (20% of y)

1️⃣ What this is

models is a Python dictionary that stores machine learning models and their hyperparameters for later use.
Each key is the name of the model, and the value is another dictionary with:

- 'model' → the model object itself
- 'params' → a dictionary of hyperparameters you want to try for grid search or tuning

2️⃣ Each model explained
---------------------------------------------------------------------------------------------
Linear Regression

'model': LinearRegression(),
'params': {}

- Linear Regression has no hyperparameters to tune here, so 'params' is empty {}.
---------------------------------------------------------------------------------------------

Decision Tree Regressor

'model': DecisionTreeRegressor(),
'params': {'max_depth': [3, 5,10], 'min_samples_split':[2,5]}

- DecisionTreeRegressor() → A decision tree model for regression.
- max_depth → maximum depth of the tree (limits complexity)
- min_samples_split → minimum number of samples required to split a node

These parameters are in lists, meaning you can try all combinations with GridSearchCV.
----------------------------------------------------------------------------------------------------

Random Forest Regressor

'model': RandomForestRegressor(),
'params': {'n_estimators': [50, 100], 'max_depth': [5,10]}

- RandomForestRegressor() → A forest of decision trees for regression.
- n_estimators → number of trees in the forest
- max_depth → maximum depth of each tree

These will also be tuned with all combinations using GridSearchCV.
=======================================================================================================

3️⃣ Why we do this

This structure makes it easy to loop through multiple models and try different hyperparameters systematically.

Example of usage:

from sklearn.model_selection import GridSearchCV

for name, m in models.items():
    grid = GridSearchCV(m['model'], m['params'], cv=5)
    grid.fit(X_train, y_train)
    print(name, grid.best_params_, grid.best_score_)

This will train all models, tune hyperparameters, and show the best settings and scores.

✅ In short:
This dictionary is a ready-to-use setup for trying multiple models and tuning their hyperparameters automatically.

```python
models  = {
    "Linear Regression":{
        'model': LinearRegression(),
        'params': {}
    },

    "Descision Tree":{
        'model': DecisionTreeRegressor(),
        'params': {'max_depth': [3, 5,10],'min_samples_split':[2,5]}
    },

    'Random Forest':{
        'model': RandomForestRegressor(),
        'params': {'n_estimators': [50, 100],'max_depth': [5,10]}
    },
}
```

```python
best_models=[]
```

for name, config in models.items():
    print(f"Training {name}")

- models.items() → gives you key-value pairs from your models dictionary.
    name → the model name (e.g., "Linear Regression")
    config → the dictionary containing 'model' and 'params'.

- print(f"Training {name}") → just tells you which model is being trained.
-----------------------------------------------------------------------------------------------------

grid = GridSearchCV(config['model'], config['params'], cv=5, scoring='neg_mean_squared_error')

- GridSearchCV → automatically tries all combinations of hyperparameters to find the best one.
- config['model'] → the ML model object (e.g., LinearRegression())
- config['params'] → the hyperparameters to try (e.g., max_depth=[3,5,10])
- cv=5 → 5-fold cross-validation (splits training data into 5 parts to validate performance)
- scoring='neg_mean_squared_error' → the metric GridSearchCV uses to decide which model is best.
-----------------------------------------------------------------------------------------------------------------------

grid.fit(X_train, y_train)

- This trains the model on your training data for all hyperparameter combinations and selects the best one.
---------------------------------------------------------------------------------------------------------------------------

y_pred = grid.predict(X_test)

- Makes predictions on the test set using the best model found by GridSearchCV.
--------------------------------------------------------------------------------------------------------------------------

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

- mean_squared_error(y_test, y_pred) → calculates average squared difference between predicted and actual values.
- np.sqrt(...) → converts it to Root Mean Squared Error (RMSE), which is easier to interpret (same units as your target).
--------------------------------------------------------------------------------------------------------------------------

r2 = r2_score(y_test, y_pred)

- r2_score → calculates the R² score, which tells how well your model explains the variance in the data.

     R² = 1 → perfect prediction
     R² = 0 → model predicts no better than the mean
-------------------------------------------------------------------------------------------------------------------------

best_models.append({
    'model': name,
    'best_params': grid.best_params_,
    'rmse': rmse,
    'r2': r2
})

- Saves the results for each model into a list best_models.
- For each model, it stores:

      - 'model' → the model name
      - 'best_params' → the hyperparameters that GridSearchCV found as best
      - 'rmse' → error on the test set
      - 'r2' → R² score on the test set
      
======================================================================================================================      

✅ In short:
This code:

1. Loops through all models in your dictionary.
2. Tunes hyperparameters with GridSearchCV.
3. Tests the best model on your test set.
4. Saves the performance metrics and best parameters in best_models.

```python
for name, config in models.items():
    print(f"Training {name}")

    grid  = GridSearchCV(config['model'], config['params'], cv=5, scoring='neg_mean_squared_error')
    grid.fit(X_train, y_train)

    y_pred = grid.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    best_models.append({
        'model': name,
        'best_params': grid.best_params_,
        'rmse': rmse,
        'r2': r2
    })
```

joblib.load('best_model.pkl').predict(X_test)

This is a one-line shortcut that does two things:

1. Load a saved model

2. Immediately use it to make predictions
----------------------------------------------------------------------------------------------------------------------------

Step 1: joblib.load('best_model.pkl')

- joblib.load() loads a previously saved model from a file.
- 'best_model.pkl' is the file that contains your trained model.
- This file was probably created earlier using:

    joblib.dump(best_model, 'best_model.pkl')

After loading, it returns the trained model object.
============================================================================================================================

Step 2: .predict(X_test)

- .predict() is a method used to make predictions.

- X_test is your test dataset (input features).

- The model uses what it learned during training to predict outputs for X_test.
============================================================================================================================

What the whole line does
   joblib.load('best_model.pkl').predict(X_test)

It means:

Load the saved model from best_model.pkl and immediately use it to predict the outputs for X_test.
============================================================================================================================

Equivalent (Long Version)

This one line is the same as writing:

  model = joblib.load('best_model.pkl')
  predictions = model.predict(X_test)

The short version just skips creating the temporary variable.
----------------------------------------------------------------------------------------------------------------------------

Why this is useful

- You don’t need to retrain the model again.
- Useful in deployment (like in web apps or APIs).
- Faster and cleaner when making predictions in production.

```python
joblib.load('best_model.pkl').predict(X_test)
```
